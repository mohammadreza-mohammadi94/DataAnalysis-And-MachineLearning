## 1.1 Factors That Would Affect On Car's Pricing <a id=2></a>

The pricing of a car can be influenced by a variety of factors, ranging from inherent characteristics of the vehicle itself to external market dynamics. Here are some specific items that can affect a car's pricing:

1. **Make and Model**: The brand and model of a car play a significant role in determining its price. Prestigious brands or luxury models often command higher prices due to factors such as brand reputation, perceived quality, and features.

2. **Vehicle Age**: Newer cars generally have higher prices compared to older ones. Depreciation is a key factor here, as cars lose value over time due to factors like wear and tear, technological obsolescence, and changing consumer preferences.

3. **Mileage**: The mileage or distance a car has traveled can impact its price. Lower mileage typically indicates less wear and tear on the vehicle, potentially leading to a higher resale value.

4. **Condition**: The overall condition of the car, including its exterior, interior, mechanical components, and maintenance history, can significantly influence its price. Well-maintained cars in good condition typically command higher prices than those with visible damage or mechanical issues.

5. **Features and Options**: Additional features and options such as leather upholstery, premium sound systems, navigation systems, advanced safety features, and upgraded wheels can increase the price of a car. Buyers are often willing to pay more for vehicles with desirable features that enhance comfort, convenience, and performance.

6. **Market Demand**: Market demand and supply dynamics can impact pricing. Factors such as consumer preferences, trends, seasonality, and regional variations in demand can affect the perceived value of a car and its selling price.

7. **Economic Factors**: Economic conditions such as inflation, interest rates, and overall consumer confidence can influence car prices. During periods of economic downturn, consumers may be more price-sensitive, leading to lower prices or increased incentives from dealerships.

8. **Government Regulations and Incentives**: Government regulations related to emissions, safety standards, and fuel efficiency can impact car prices by influencing manufacturing costs and consumer preferences. Additionally, government incentives such as tax credits or rebates for electric or hybrid vehicles can affect pricing.

9. **Dealer Markup and Negotiation**: Dealerships often apply a markup or profit margin to the manufacturer's suggested retail price (MSRP) of a car. The final selling price can also be influenced by negotiation between the buyer and seller, as well as any discounts, promotions, or incentives offered by the dealership.

10. **Resale Value**: The anticipated resale value of a car can influence its initial pricing. Cars with higher resale values are often priced higher initially, as buyers are willing to pay more for vehicles that retain their value well over time.

Overall, the pricing of a car is influenced by a complex interplay of factors related to the vehicle itself, market conditions, consumer preferences, and economic considerations. Understanding these factors can help buyers and sellers make informed decisions when buying or selling a car.

## 5.2 Outliers Imputation Using Median <a id=26></a>

**NOTE**:  
**When you have a limited dataset and removing outliers results in a significant loss of data, it's essential to consider alternative approaches to handle outliers without discarding too much information. 
: Instead of removing outliers, you can impute them with a reasonable value. For numerical features, you can replace outliers with the median or mean of the respective feature. For categorical features, you can impute outliers with the mode.**

```python
# below funtion will find outliers by using IQR:
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))
```

```python
# Apply the function to detect outliers
outliers_mask = detect_outliers_iqr(df.select_dtypes(include=['number']))

# Replace outliers with median
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Imputation
for col in numeric_cols:
    median_val = df[col].median()  # Calculate median for each column
    df[col] = df[col].where(~outliers_mask[col], median_val)
```

### 6.1.2 More About Fuel & Engine Types <a id=31></a>

**Engine Types**:  

1. **OHC (Overhead Camshaft)**:
   - OHC engines have the camshaft(s) mounted above the cylinder head, which operates the valves directly or through short rocker arms. This design allows for better engine performance, efficiency, and higher revving capability compared to older designs like pushrod engines.

2. **OHCF (Overhead Camshaft, Front)**:
   - OHCF engines are similar to OHC engines but typically have a front-mounted camshaft configuration. This design is commonly used in modern engines and offers advantages in terms of compactness, weight reduction, and improved valve control.

3. **OHCV (Overhead Camshaft, V-shaped)**:
   - OHCV engines have an overhead camshaft configuration arranged in a V-shaped cylinder block. This design is often found in V-shaped engines, such as V6 or V8 configurations, and allows for better packaging of the engine components within the engine bay.

4. **DOHC (Double Overhead Camshaft)**:
   - DOHC engines have two camshafts per cylinder bank, one for intake valves and one for exhaust valves. This design allows for precise control of valve timing and enables higher engine performance compared to single overhead camshaft designs.

5. **L (L-Type Engine)**:
   - L-type engines may refer to various engine designs or configurations depending on the specific context. Without further information, it's challenging to provide a precise explanation for this designation.

6. **Rotor (Rotary Engine)**:
   - Rotor engines, also known as rotary engines, are a type of internal combustion engine that operates on a circular motion of a triangular rotor within a chamber. This design offers advantages in terms of simplicity, compactness, and smooth operation but is less commonly used in mainstream automotive applications.

7. **DOHCV (Double Overhead Camshaft, V-shaped)**:
   - DOHCV engines combine the features of DOHC and OHCV designs, featuring two camshafts per cylinder bank arranged in a V-shaped cylinder block. This configuration offers the benefits of precise valve control and efficient packaging fautomotive engineering.

**Fuel System**:  

1. **MPFI (Multi-Point Fuel Injection)**:
   - MPFI is a fuel injection system that injects fuel into multiple intake ports or cylinders simultaneously. It delivers fuel more precisely to each cylinder, optimizing engine performance, fuel efficiency, and emissions control. MPFI systems are commonly found in modern gasoline-powered vehicles.

2. **2BBL (Two-Barrel Carburetor)**:
   - 2BBL refers to a carburetor with two separate barrels or venturis, through which air and fuel mixture flow into the engine's intake manifold. This carburetor design is relatively simple and cost-effective but offers less precise fuel metering compared to fuel injection systems.

3. **IDI (Indirect Injection)**:
   - IDI is a type of diesel fuel injection system where fuel is injected into a pre-combustion chamber or swirl chamber before entering the main combustion chamber of the engine. It is characterized by its indirect injection process, which helps improve fuel atomization and combustion efficiency in diesel engines.

4. **1BBL (One-Barrel Carburetor)**:
   - 1BBL refers to a carburetor with a single barrel or venturi, through which air and fuel mixture flow into the engine's intake manifold. This carburetor design is simpler and less efficient than multi-barrel carburetors but was commonly used in older vehicles.

5. **SPDI (Sequential Port Fuel Injection)**:
   - SPDI is a fuel injection system that sequentially injects fuel into each intake port or cylinder individually, based on the engine's firing order. This sequential injection method enhances fuel distribution, throttle response, and engine efficiency compared to non-sequential systems.

6. **4BBL (Four-Barrel Carburetor)**:
   - 4BBL refers to a carburetor with four separate barrels or venturis, providing increased airflow and fuel delivery capacity compared to two-barrel carburetors. Four-barrel carburetors are often used in high-performance engines to maximize power output.

7. **MFI (Multi-Point Fuel Injection)**:
   - MFI is similar to MPFI and refers to a fuel injection system that injects fuel into multiple intake ports or cylinders simultaneously. It offers the same benefits of precise fuel delivery and engine performance optimization.

8. **SPFI (Single-Point Fuel Injection)**:
   - SPFI is a fuel injection system that injects fuel into a single point or location in the intake manifold, where it mixes with incoming air before entering the engine's cylinders. While simpler and less expensive than multi-point systems, SPFI may offer lower performance and efficiency.
